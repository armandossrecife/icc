```python
import pandas as pd
```

```python
def checa_alunos_e_listas(lista_nomes, lista_nomes_lista1, lista_nomes_lista2, lista_nomes_lista3, lista_nomes_lista4):
  # Checa os alunos que entregaram no prazo (postado nas tarefas via SIGAA) as listas l1, l2, l3 e l4
  fez_l1, fez_l2, fez_l3, fez_l4 = [], [], [], []
  for aluno in lista_nomes:
    if aluno in lista_nomes_lista1:
      fez_l1.append(1)
    else:
      fez_l1.append(0)
    if aluno in lista_nomes_lista2:
      fez_l2.append(1)
    else:
      fez_l2.append(0)
    if aluno in lista_nomes_lista3:
      fez_l3.append(1)
    else:
      fez_l3.append(0)
    if aluno in lista_nomes_lista4:
      fez_l4.append(1)
    else:
      fez_l4.append(0)
  return fez_l1, fez_l2, fez_l3, fez_l4

def cria_dicionario_alunos_listas(l1, l2, l3, l4):
  dict_listas_av2 = {
      'fez_l1':l1, 'fez_l2':l2, 'fez_l3':l3, 'fez_l4':l4
  }
  return dict_listas_av2
```

```python
# importa dados para pre-analise da av2
!wget https://raw.githubusercontent.com/armandossrecife/icc/main/alunosicc.csv
!wget https://raw.githubusercontent.com/armandossrecife/icc/main/respostas_lista1sr.csv
!wget https://raw.githubusercontent.com/armandossrecife/icc/main/respostas_lista2.csv
!wget https://raw.githubusercontent.com/armandossrecife/icc/main/respostas_lista3sr.csv
!wget https://raw.githubusercontent.com/armandossrecife/icc/main/respostas_lista4sr.csv
```

```python
# Ler arquivos .csv
df_alunos = pd.read_csv('alunosicc.csv')
df_alunos_lista1 = pd.read_csv('respostas_lista1sr.csv', sep=';')
df_alunos_lista2 = pd.read_csv('respostas_lista2.csv', sep=';')
df_alunos_lista3 = pd.read_csv('respostas_lista3sr.csv', sep=';')
df_alunos_lista4 = pd.read_csv('respostas_lista4sr.csv', sep=';')

# Carrega as listas de nomes e as listas da av2
lista_nomes = df_alunos['nome'].tolist()
lista_nomes_lista1 = df_alunos_lista1.Nome.tolist() 
lista_nomes_lista2 = df_alunos_lista2.Nome.tolist()
lista_nomes_lista3 = df_alunos_lista3.Nome.tolist()
lista_nomes_lista4 = df_alunos_lista4.Nome.tolist()

# Garante que os nomes estao ordenados
lista_nomes = sorted(lista_nomes)
lista_nomes_lista1 = sorted(lista_nomes_lista1)
lista_nomes_lista2 = sorted(lista_nomes_lista2)
lista_nomes_lista3 = sorted(lista_nomes_lista3)
lista_nomes_lista4 = sorted(lista_nomes_lista4)
```

```python
l1, l2, l3, l4 = checa_alunos_e_listas(lista_nomes, lista_nomes_lista1, lista_nomes_lista2, lista_nomes_lista3, lista_nomes_lista4)
dict_listas_av2 = cria_dicionario_alunos_listas(l1, l2, l3, l4)

df_listas_av2 = pd.DataFrame(dict_listas_av2)

df_alunos[['fez_l1', 'fez_l2', 'fez_l3', 'fez_l4']] = df_listas_av2[['fez_l1', 'fez_l2', 'fez_l3', 'fez_l4']]

df_alunos['qtd_listas_entregues'] = df_alunos.fez_l1 + df_alunos.fez_l2 + df_alunos.fez_l3 + df_alunos.fez_l4
```
