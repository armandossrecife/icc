## Exercício (Manipulando arquivos)

Crie um programa em python que manipula arquivos para leitura e escrita.

Dado o link https://raw.githubusercontent.com/armandossrecife/icc/main/municipios.csv correspondente a lista de municípios registrados pelo TSE nas eleições 2022.

Utilizando as funções de manipulação de arquivos do Python. Gere o seguinte relatório:

1. Total de municípios registrados.

2. Mostre a quantidade total de município para cada estado. 

3. Liste apenas os municípios do estado do Piauí (PI).

4. Liste os municípios do Piauí (PI) em ordem alfabética crescente.

5. Liste os municípios do Piauí (PI) em ordem alfabética decrescente

6. Crie os seguintes arquivos:

6.1. Crie um novo arquivo, chamado piaui.csv, contendo seus municípios

6.2. Crie um novo arquivo, chamado estados.csv, contendo as siglas dos estados e a quantidade total de municípios de cada estado.

```python
import requests
import csv

# baixa o arquivo da url usando o get da biblioteca requests
url = 'https://raw.githubusercontent.com/armandossrecife/icc/main/municipios.csv'
resposta = requests.get(url)

# Salva o conteudo no arquivo local municipios.csv
with open('municipios.csv', 'wb') as arquivo_baixado:
    arquivo_baixado.write(resposta.content)

# Abre o arquivo CSV contendo os municipios
with open('municipios.csv') as arquivo:
    leitor_csv = csv.reader(arquivo, delimiter=';')
    
    # Inicializa as variáveis
    total_municipios = 0
    municipios_por_estado = {}
    municipios_pi = []
    
    # Itera pelas linhas do arquivo CSV
    for linha in leitor_csv:
        uf = linha[0]
        codigo = linha[1]
        nome = linha[2]
        
        # Atualiza as variáveis
        total_municipios += 1
        municipios_por_estado[uf] = municipios_por_estado.get(uf, 0) + 1
        
        # Salva os municipios do PI em uma lista
        if uf == 'PI':
            municipios_pi.append(nome)
    
    # Exibe o total de municípios registrados
    print(f'Total de municípios registrados: {total_municipios-1}')
    
    # Exibe a quantidade de municípios por estado
    print('Quantidade de municípios por estado:')
    for uf, quantidade in municipios_por_estado.items():
        print(f'{uf}: {quantidade}')
    
    # Exibe apenas os municípios do estado do Piauí (PI)
    print('Municípios do estado do Piauí (PI):')
    for municipio in municipios_pi:
        print(municipio)
        
    # Cria o arquivo piaui.csv contendo os municípios do estado do Piauí (PI)
    with open('piaui.csv', 'w', newline='') as arquivo_piaui:
        escritor_csv = csv.writer(arquivo_piaui)
        for municipio in municipios_pi:
            escritor_csv.writerow([municipio])
    
    # Cria o arquivo estados.csv contendo a quantidade de municípios por estado
    with open('estados.csv', 'w', newline='') as arquivo_estados:
        escritor_csv = csv.writer(arquivo_estados)
        for uf, quantidade in municipios_por_estado.items():
            escritor_csv.writerow([uf, quantidade])
```

```python
# Abre o arquivo CSV contendo os municipios
with open('municipios.csv') as arquivo:
    leitor_csv = csv.reader(arquivo, delimiter=';')
    
    # Inicializa as variáveis
    municipios_pi = []
    
    # Itera pelas linhas do arquivo CSV
    for linha in leitor_csv:
        uf = linha[0]
        codigo = linha[1]
        nome = linha[2]
                
        # Salva os municipios do PI em uma lista
        if uf == 'PI':
            # remove acentos e caracteres especiais
            nome = unidecode(nome)
            municipios_pi.append(nome)
            
    # Exibe apenas os municípios do estado do Piauí (PI)
    print('Municípios do estado do Piauí (PI):')
    for municipio in municipios_pi:
        print(municipio)
    
    # Exibe os municípios do Piauí (PI) em ordem alfabética crescente
    print('Municípios do estado do Piauí (PI) em ordem alfabética crescente:')
    for municipio in sorted(municipios_pi):
        print(municipio)
    
    # Exibe os municípios do Piauí (PI) em ordem alfabética decrescente
    print('Municípios do estado do Piauí (PI) em ordem alfabética decrescente:')
    for municipio in sorted(municipios_pi, reverse=True):
        print(municipio)
    
    # Cria o arquivo piaui.csv contendo os municípios do estado do Piauí (PI)
    with open('piaui.csv', 'w', newline='') as arquivo_piaui:
        escritor_csv = csv.writer(arquivo_piaui)
        for municipio in municipios_pi:
            escritor_csv.writerow([municipio])
```
