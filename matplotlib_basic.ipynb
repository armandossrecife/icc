### Propriedades mais importantes

Vamos começar criando um gráfico simples e, em seguida, discutiremos as diferentes propriedades que você pode usar para personalizá-lo.

```python
# import as bibliotecas matplotlib e numpy
import matplotlib.pyplot as plt
import numpy as np

# Criar um conjunto de dados de exemplo
# https://numpy.org/doc/stable/reference/generated/numpy.linspace.html
x = np.linspace(0, 10, 100)
print(f'x: {x}')

# https://numpy.org/doc/stable/reference/generated/numpy.sin.html
y = np.sin(x)
print(f'y: {y}')

# Plotar os dados
plt.plot(x, y)

# Mostrar o gráfico
plt.show()
```

Este código cria um gráfico de uma função senoidal simples e mostra-o na tela. 

Agora, vamos explorar as diferentes propriedades que você pode usar para personalizar esse gráfico.

**Eixo x** e **eixo y**

Os eixos x e y são essenciais em qualquer gráfico e indicam as dimensões de seus dados. 

Você pode personalizar esses eixos para mostrar intervalos específicos ou números de etiquetas.

```python
# Personalizar os eixos
plt.xlim([0, 10])
plt.ylim([-1, 1])

# Mostrar o gráfico
plt.show()
```

A função xlim() define os limites do eixo x e a função ylim() define os limites do eixo y. 

No exemplo acima, definimos o intervalo do eixo x para [0, 10] e o intervalo do eixo y para [-1, 1]. 

Isso significa que o eixo x vai de 0 a 10 e o eixo y de -1 a 1.

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plotar os dados
plt.plot(x, y)
plt.xlim([0, 10])
plt.ylim([-1, 1])

# Mostrar o gráfico
plt.show()
```

**xlabel** e **ylabel**

Os rótulos dos eixos x e y informam ao leitor o que está sendo medido em cada eixo. 

Você pode personalizar esses rótulos usando as funções xlabel() e ylabel().

```python
# Personalizar os rótulos dos eixos
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')

# Mostrar o gráfico
plt.show()
```

No exemplo acima, adicionamos rótulos aos eixos x e y. O rótulo do eixo x é "Tempo (s)" e o rótulo do eixo y é "Amplitude".

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plotar os dados
plt.plot(x, y)
plt.xlim([0, 10])
plt.ylim([-1, 1])

# Personalizar os rótulos dos eixos
plt.xlabel('x')
plt.ylabel('y')

# Mostrar o gráfico
plt.show()
```

**title**

O título do gráfico é uma descrição geral do que está sendo representado no gráfico. 

Você pode personalizar o título usando a função title().

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plotar os dados
plt.plot(x, y)
plt.xlim([0, 10])
plt.ylim([-1, 1])

# Personalizar o título do gráfico
plt.title('Função Senoidal')
# Personalizar os rótulos dos eixos
plt.xlabel('x')
plt.ylabel('y')

# Mostrar o gráfico
plt.show()
```

**plot**

A função plot() é usada para plotar os dados em um gráfico. Você pode personalizar o estilo do gráfico adicionando argumentos à função plot(), como a cor e o estilo da linha.

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Personalizar o estilo do gráfico
plt.plot(x, y, color='red', linestyle='dashed')
plt.xlim([0, 10])
plt.ylim([-1, 1])

# Personalizar o título do gráfico
plt.title('Função Senoidal')
# Personalizar os rótulos dos eixos
plt.xlabel('x')
plt.ylabel('y')


# Mostrar o gráfico
plt.show()
```

**grid**

O grid é uma grade que pode ser adicionada a um gráfico para ajudar na leitura dos dados. É uma grade de linhas que se estendem ao longo dos eixos x e y, permitindo que você visualize facilmente os pontos de dados em relação a essas linhas.



Para adicionar um grid a um gráfico do Matplotlib, você pode usar a função grid(). Essa função aceita vários argumentos, incluindo b (um valor booleano que determina se o grid deve ser exibido ou oculto), which (que especifica os rótulos dos eixos que devem ter um grid - 'major' para os principais rótulos, 'minor' para os rótulos secundários, ou 'both' para ambos), axis (que especifica qual eixo deve ter o grid - 'x' para o eixo x, 'y' para o eixo y, ou 'both' para ambos) e outros argumentos relacionados à aparência do grid, como a cor e o estilo das linhas.

```python
import matplotlib.pyplot as plt
import numpy as np

# Criar um conjunto de dados de exemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plotar os dados com um grid
plt.plot(x, y)
plt.grid(b=True, which='both', axis='both', color='gray', linestyle='--', linewidth=0.5)

# Personalizar o título e os rótulos dos eixos
plt.title('Função Senoidal')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')

# Mostrar o gráfico
plt.show()
```

Neste exemplo, adicionamos um grid ao gráfico usando a função grid(). Definimos o argumento b como True para mostrar o grid, o argumento which como 'both' para adicionar grid a ambos os rótulos do eixo x e y, e o argumento axis também como 'both' para adicionar o grid a ambos os eixos. Definimos a cor do grid como cinza (color='gray'), o estilo das linhas como tracejado (linestyle='--') e a largura das linhas como 0.5 pontos (linewidth=0.5).

### Gráficos mais comuns

Gráfico de uma reta (função do 1o. grau)

```python
# Define the equation of the line
def f(x):
    return 2*x + 1

# Create an array of x values
x = np.linspace(-5, 5, 100)

print(f'qtd: {len(x)}, x: {x}')

# Evaluate the equation of the line for each x value
y = f(x)
print(f'qtd: {len(y)}, y: {y}')

# Plot the line
plt.plot(x, y)

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Equation: y = 2x + 1')

# Display the plot
plt.show()
```

Gráfico de uma curva (função do 2o. grau)

```python
# Define the equation of the curve
def f(x):
    return x**2 - 2*x + 1

# Create an array of x values
x = np.linspace(-5, 7, 100)
print(f'qtd: {len(x)}, x: {x}')

# Evaluate the equation of the curve for each x value
y = f(x)
print(f'qtd: {len(y)}, y: {y}')

# Plot the curve
plt.plot(x, y)

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic Equation: y = x^2 - 2x + 1')

# Display the plot
plt.show()
```

Gráfico de uma curva (função de 3o. grau - função polinomial)

```python
# Define the equation of the function
def f(x):
    return x**3 - 6*x**2 + 11*x - 6

# Create an array of x values
x = np.linspace(-5, 5, 100)

# Evaluate the equation of the function for each x value
y = f(x)

# Plot the function
plt.plot(x, y)

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomial Function: y = x^3 - 6x^2 + 11x - 6')

# Display the plot
plt.show()
```

Gráfico de Barras

```python
# Create some data
# categorias (A, B, C, D e E)
x = ['A', 'B', 'C', 'D', 'E']

# valores de cada categoria 
y = [10, 24, 36, 45, 52]

# Create a bar plot
plt.bar(x, y)

# Add labels and title
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Bar Plot')

# Display the plot
plt.show()
```

Gráfico de Pizza (Pie chart)

```python
# Dados de exemplo
labels = ['Maçãs', 'Laranjas', 'Bananas', 'Pêssegos']
valores = [25, 30, 35, 10]

# Criar o gráfico de pizza
plt.pie(valores, labels=labels)

# Adicionar um título
plt.title('Frutas Favoritas')

# Mostrar o gráfico
plt.show()
```

Gráfico de dipersão (scatter plot)

O gráfico de dispersão é uma ferramenta gráfica amplamente utilizada em estatística e outras áreas do conhecimento para visualizar a relação entre duas variáveis quantitativas. Ao plotar os pontos no gráfico, é possível detectar padrões e tendências nos dados e identificar possíveis correlações entre as variáveis.

https://en.wikipedia.org/wiki/Scatter_plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some data
x = np.random.rand(100)
y = np.random.rand(100)

# Create a scatter plot
plt.scatter(x, y)

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot')

# Display the plot
plt.show()
```

Histograma

https://en.wikipedia.org/wiki/Histogram

```python
import matplotlib.pyplot as plt
import numpy as np

# Gerar dados de exemplo
dados = np.random.randn(1000)

# Criar o histograma
plt.hist(dados, bins=10)

# Adicionar rótulos e título
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.title('Histograma de dados')

# Mostrar o histograma
plt.show()
```

Boxsplot

https://en.wikipedia.org/wiki/Box_plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Gerar dados de exemplo
dados1 = np.random.normal(0, 1, 100)
dados2 = np.random.normal(1, 1, 100)
dados3 = np.random.normal(2, 1, 100)

# Colocar dados em uma lista
dados = [dados1, dados2, dados3]

# Criar o boxplot
plt.boxplot(dados)

# Adicionar rótulos e título
plt.xticks([1, 2, 3], ['Dados 1', 'Dados 2', 'Dados 3'])
plt.ylabel('Valores')
plt.title('Boxplot de dados')

# Mostrar o boxplot
plt.show()
```

### Usando dados de arquivos .csv

Gráfico de barras

```python
import pandas as pd

gdp_by_country = 'https://raw.githubusercontent.com/armandossrecife/teste/main/data_country.csv'

# Load the data from the .csv file into a pandas DataFrame
df = pd.read_csv('https://raw.githubusercontent.com/armandossrecife/teste/main/data_country.csv')

# Create a bar plot
plt.bar(df['Country'], df['GDP(trillions)'])

# Add labels and a title to the plot
plt.xlabel('Country')
plt.ylabel('GDP (in trillions)')
plt.title('GDP by Country')

# Show the plot
plt.show()
```

```python
# Load the data from the .csv file into a pandas DataFrame
df = pd.read_csv(gdp_by_country)

# Create a bar plot
plt.bar(df['Country'], df['GDP(trillions)'])

# rotate the x-axis labels by 90 degrees
plt.xticks(rotation=90)

# Add labels and a title to the plot
plt.xlabel('Country')
plt.ylabel('GDP (in trillions)')
plt.title('GDP by Country')

# Show the plot
plt.show()
```

```python
# Load the data from the .csv file into a pandas DataFrame
df = pd.read_csv(gdp_by_country)

# Create a bar plot
plt.barh(df['Country'], df['GDP(trillions)'])

# Add labels and a title to the plot
plt.xlabel('GDP (in trillions)')
plt.ylabel('Country')

plt.title('GDP by Country')

# Show the plot
plt.show()
```
