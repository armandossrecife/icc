```python
import requests

url_arquivo =  'https://raw.githubusercontent.com/armandossrecife/icc/main/municipios.csv'

#faz download do arquivo municipios.csv
resposta = requests.get(url_arquivo)

f_meu_arquivo = open('cidades.csv', 'wb')
f_meu_arquivo.write(resposta.content)
f_meu_arquivo.close()
print('Arquivo cidades.csv criado com sucesso!')
```

# Quantas cidades do PI foram registradas nas eleições 2022?

```python
f_arquivo = open("cidades.csv", 'r')
lista_cidades_pi = []

for linha in f_arquivo:
  if 'PI;' in linha:
    linha = linha.strip()
    elemento = linha.split(';')
    cidade = elemento[2]
    lista_cidades_pi.append(cidade)

f_arquivo.close()
```

```python
print(f'{len(lista_cidades_pi)} foram registradas nas eleições 2022.')
```

```python
for cidade in lista_cidades_pi:
  print(cidade)
```

# Crie uma lista contendo apenas os estados registrados nas eleições 2022

```python
f_arquivo = open('cidades.csv', 'r')

lista_estados = []

for linha in f_arquivo:
  elemento = linha.split(';')
  estado = elemento[0]
  lista_estados.append(estado)

f_arquivo.close()

conjunto_estado = set(lista_estados)
print(f'Conjunto de estados: {conjunto_estado}')

lista_estados_unicos = list(conjunto_estado)
print(f'Lista de estados: {lista_estados_unicos}')
```

# Mostre as cidades agrupadas por estado

```python
dict_estados = {}

print(lista_estados_unicos)
```

```python
lista_cidades = []

for estado in lista_estados_unicos:
  f_arquivo = open('cidades.csv', 'r')
  for linha in f_arquivo:
    elemento = linha.split(';')
    uf = elemento[0]
    cidade = elemento[2].strip()
    if estado==uf: 
      lista_cidades.append(cidade)
  f_arquivo.close()
  dict_estados[estado] = lista_cidades
  lista_cidades = []
```
